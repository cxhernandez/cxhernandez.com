````python
from msmbuilder import tpt
import matplotlib.pyplot as pp
import numpy as np
import seaborn as sns
import networkx as nx
import mpld3
````

````python
N = 25
M = np.random.binomial(10,.01,size=(N,N))
````

````python
M2 = np.zeros((N,N))
for i,c in enumerate(M):
    M2[:,i] = np.array(map(float,c))/sum(c)
````

````python
G = nx.Graph(data=M2)
````

````python
sns.set_style('dark')
mpld3.enable_notebook()
fig = pp.figure(figsize=(8,8))
ax  = fig.add_subplot(1, 1, 1)
# with nodes colored by degree sized by population
node_color=np.array([float(len(G[v])) for v in G])
pos = nx.spring_layout(G)
pos_array = np.array(pos.values()).T
sizes = 5000*np.array(nx.pagerank_scipy(G).values())
for i in G:
    for j in G[i]:
        ax.plot([pos_array[0,i],pos_array[0,j]],[pos_array[1,i],pos_array[1,j]],'-w')
pts = ax.scatter(pos_array[0],pos_array[1],sizes, color = cm.coolwarm(node_color/node_color.max()),zorder=10)
pp.xticks([])
pp.yticks([])
        
mpld3.plugins.connect(fig,mpld3.plugins.PointLabelTooltip(pts,pos.keys()))
````

````python
from msmbuilder import tpt
import matplotlib.pyplot as pp
import numpy as np
import seaborn as sns
import networkx as nx
import mpld3
````

````python
N = 25
M = np.random.binomial(10,.01,size=(N,N))
````

````python
M2 = np.zeros((N,N))
for i,c in enumerate(M):
    M2[:,i] = np.array(map(float,c))/sum(c)
````

````python
G = nx.Graph(data=M2)
````

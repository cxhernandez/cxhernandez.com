````python
%pylab inline
from msmbuilder import tpt
import numpy as np
from scipy.stats import poisson
import networkx as nx
import seaborn, mpld3
````

````python
#from http://stackoverflow.com/questions/6703792/how-does-random-graph-generator-work-in-networkx
def poissongraph(n,mu):
    z= np.zeros(n)
    for i in range(n):
        z[i]=poisson.rvs(mu)
    G=nx.expected_degree_graph(z).to_directed()
    for j,item in enumerate(G.adjacency_list()):
        w = np.random.randint(low=1,high=100,size=len(item))
        W = w.sum()
        for i in item:
            G[j][i]['weight'] = w[i]/W
        
    return G
````
